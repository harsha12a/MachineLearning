```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = pd.read_csv('kmeans.csv')
dataset.head()
```

```python
x = dataset.iloc[:, [3, 4]].values
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmn = KMeans(n_clusters=i, init='k-means++')
    kmn.fit(x)
    wcss.append(kmn.inertia_)
plt.plot(range(1, 11), wcss)
```

```python
kmeans = KMeans(n_clusters=5, init='k-means++')
kmeans.fit(x)
y_pred = kmeans.predict(x)
```

```python
plt.plot(x[y_pred == 0, 0], x[y_pred == 0, 1], 'ro', label='Cluster 1')
plt.plot(x[y_pred == 1, 0], x[y_pred == 1, 1], 'go', label='Cluster 2')
plt.plot(x[y_pred == 2, 0], x[y_pred == 2, 1], 'bo', label='Cluster 3')
plt.plot(x[y_pred == 3, 0], x[y_pred == 3, 1], 'mo', label='Cluster 4')
plt.plot(x[y_pred == 4, 0], x[y_pred == 4, 1], 'co', label='Cluster 5')
plt.plot(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], 'yo', label='Centroids')
plt.xlabel('Annual Income')
plt.legend()
```
